# Pie (Treemap)

```python
import matplotlib.pyplot as plt
import squarify

# Dados
labels = ['Categoria A', 'Categoria B', 'Categoria C', 'Categoria D']
sizes = [20, 35, 25, 20]

# Cria o gr√°fico
squarify.plot(sizes=sizes, label=labels, color=["red","green","blue","yellow"])
plt.axis('off')
plt.show()
```
