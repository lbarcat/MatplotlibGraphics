# 3D Graph

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Cria o gráfico
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Dados
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
z = [0, 1, 2, 3, 4]

# Plota os dados
ax.scatter(x, y, z)

# Exibe o gráfico
plt.show()
```
